# Grids

This notebook shows how to produce **grids** using **consav.grids**.

# Setup

```python
%load_ext autoreload
%autoreload 2

import time
import numpy as np

import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
prop_cycle = plt.rcParams["axes.prop_cycle"]
colors = prop_cycle.by_key()["color"]
import ipywidgets as widgets

# load the module
from consav import grids
```

# Settings

```python
ns = [5,10,15,20,25,30]
phis = [1,1.5,2.0]
```

# `nonlinspace`

**Illustration:**

```python
for phi in phis:
    
    fig = plt.figure()
    ax = fig.add_subplot(1,1,1)
    ax.set_title(f'phi = {phi}')

    for n in ns:
        x = grids.nonlinspace(0,10,n,phi)
        ax.scatter(n*np.ones(n),x)
```

# `equilogspace`

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.set_title(f'equilogspace')

for n in ns:
    x = grids.equilogspace(0,10,n)
    ax.scatter(n*np.ones(n),x)
```
